# Plotting and Data Visualization

```python
#Prepare data for plotting

import os 
import numpy

distance_file = os.path.join('data', 'distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')

headers = distances[0]
data = distances[1:]
data = data.astype(numpy.float)
print (data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(data[:,1])
```

```python
sample_name = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
#plt.plot(data[:,1])
plt.plot(data[:,1], label=sample_name)
plt.legend() #adds the sample label 

plt.savefig(F'{sample_name}.png', dpi=300) #use default dpi initially. For higher resolution, mention dpi
```

```python
#plot two sets of data
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.plot(data[:, 1], label=headers[1])
plt.plot(data[:, 2], label=headers[2])
plt.legend
plt.savefig ('two_samples.png')
```

```python
plt.figure()
for col in range(1, len(headers)):
    plt.plot(data[:, col], label=headers[col])

plt.legend()
plt.xlabel('Simulation Frame') #outside for loop since we want to do it only once
plt.ylabel('Distance (angstrom)')
plt.savefig('all_sample.png')    
```

```python
#Make a plot for each set of data move everything into for loop

for col in range(1, len(headers)):
    plt.figure()
    plt.plot(data[:, col], label=headers[col])
    plt.legend()
    plt.xlabel('Simulation Frame') 
    plt.ylabel('Distance (angstrom)')
    name = headers[col]
    plt.savefig(F'{name}.png') 
```
